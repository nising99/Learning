```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def functionA(numberOfDice, numberOfDiceNext, y):
  for i in range(numberOfDice):
    diceNumber = random.randint(1, 6)
    if diceNumber == 6:
      numberOfDiceNext = numberOfDiceNext - 1
  y.append(numberOfDiceNext)
  if numberOfDiceNext > 0:
    functionA(numberOfDiceNext, numberOfDiceNext, y)

num = 100000

y = [num]
#print(y)
functionA(num, num, y)
#print(y)
#print(len(y))

x = []
for j in range(len(y)):
  x.append(j)
#print(x)

xpoints = np.array(x)
ypoints = np.array(y)

plt.plot(xpoints, ypoints)
plt.show()
```
